```python
import pandas as pd
import numpy as np
```

# 1) Internet users data:


```python
# Code_1
df = pd.read_csv("InternetUsersDataset.csv")
df.head()
```

```python
# Code_2
df_melt = df.melt(id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
             var_name='Year',
             value_name = 'Internet users (per 100 people)')

df_drop = df_melt.dropna()
df_drop
```

```python
df_drop.head(1)
```

```python
# Code_3
df_21 = df_drop[['Country Name', 'Country Code', 'Year', 'Internet users (per 100 people)']]
```

```python
df_21.head()
```

```python
# Code_4
df_21st = df_21[df_21['Year']>='2000']
```

```python
df_21st.head()
```

# Yearly Internet users (per 100 people)

```python
# Code_5
df_2000 = df_21st[df_21st['Year']=='2000']
df_2001 = df_21st[df_21st['Year']=='2001']
df_2002 = df_21st[df_21st['Year']=='2002']
df_2003 = df_21st[df_21st['Year']=='2003']
df_2004 = df_21st[df_21st['Year']=='2004']
df_2005 = df_21st[df_21st['Year']=='2005']
df_2006 = df_21st[df_21st['Year']=='2006']
df_2007 = df_21st[df_21st['Year']=='2007']
df_2008 = df_21st[df_21st['Year']=='2008']
df_2009 = df_21st[df_21st['Year']=='2009']
df_2010 = df_21st[df_21st['Year']=='2010']
df_2011 = df_21st[df_21st['Year']=='2011']
df_2012 = df_21st[df_21st['Year']=='2012']
df_2013 = df_21st[df_21st['Year']=='2013']
df_2014 = df_21st[df_21st['Year']=='2014']
df_2015 = df_21st[df_21st['Year']=='2015']
df_2016 = df_21st[df_21st['Year']=='2016']
```

```python
# Code_6
df_pivot = pd.pivot_table(df_21st, values='Internet users (per 100 people)',
                          index = ['Country Name', 'Country Code'])

df_pivot = df_pivot.reset_index()


df_pivot
```

```python
# Code_7
df_pivot = df_pivot.rename(columns={'Internet users (per 100 people)':'Avergae Internet users (per 100 people)'})
df_pivot.head()
```

# 2) GDP data:

```python
# Code_8
from pandas.io.html import read_html
page = "https://en.wikipedia.org/wiki/List_of_countries_by_real_GDP_growth_rate"

wikitables = read_html(page,  attrs={"class":"wikitable"})

print ("Extracted {num} wikitables".format(num=len(wikitables)))
```

There are two Wikitables on Wikipedia page,ut we need the second one.

```python
# Code_9: To scrape the second Wikitable
Gdp = wikitables[1]
Gdp.head()
```

```python
# Code_10
gdp = Gdp.rename(columns={'Country':'Country Name',
           'Real GDP growthrate (%) 2013':'2013',
           'Real GDP growthrate (%) 2014':'2014',
           'Real GDP growthrate (%) 2015':'2015',
           'Real GDP growthrate (%) 2016':'2016',
           'Real GDP growthrate (%) 2017':'2017',
           'Real GDP growthrate (%) 2018[3]':'2018'})
```

```python
gdp.head()
```

```python
gdp.isna().sum()
```

```python
# Code_11
Gdp_average = gdp[['Country Name', 'Average GDP growthrate (%) 2013–2018']]
Gdp_average
```

```python
# Code_12
gdp_melt = gdp.melt(id_vars = ['Country Name', 'Average GDP growthrate (%) 2013–2018'],
                    var_name='Year',
                    value_name='Real GDP growthrate (%)'
                   )
gdp_melt.head()
```

# 3) Merge

# A

```python
# Code_13
merge = pd.merge(df_21, gdp_melt, on =['Country Name', 'Year'])
```

```python
merge
```

# B

```python
# Code_14
merge_average = pd.merge(df_pivot, Gdp_average, on='Country Name')
merge_average
```

# Visulaisation_2

```python
tr_in = go.Scatter(
    x= merge['Country Name'],
    y= merge['Internet users (per 100 people)'],
    name = 'Internet users (per 100 people) since 2000')

tr_gdp = go.Scatter(
    x = merge['Country Name'],
    y = merge['Real GDP growthrate (%)'],
    name= 'GDP growthrate (%) 2013–2018')

data = [tr_in, tr_gdp]

layout = go.Layout(
    title='GDP growth rate vs Internet users',
    xaxis=dict(
        title='Country Name',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
            )
        ),
    yaxis=dict(
        title='Internet users & GDP growth rate',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
            )
        )
    )


fig = go.Figure(data=data, layout = layout)

py.iplot(fig, filename='GDP growth rate vs Internet users')
```
